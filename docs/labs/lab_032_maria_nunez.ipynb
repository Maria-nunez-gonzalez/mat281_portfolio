1. Lo primero será juntar toda la información en un _solo archivo_, para ello necesitamos seguir los siguientes pasos:

 * a) Crear el archivo **df_anio**, que contenga la información de **libertad_prensa_anio.csv** para cada año. Luego, normalice el nombre de las columnas a minúscula.

> **Hint**: Para juntar por _anio_ ocupe la función **pd.concat**. Para juntar información por columna ocupe **pd.merge**.

```python
# a)

lista = []

for name in archivos_anio:
    df_temporal = pd.read_csv(name)
    lista.append(df_temporal)

df_anio = pd.concat(lista)
df_anio.columns=df_anio.columns.str.lower()
df_anio

# b)

df_codigos['codigo_iso'].value_counts()
df_codigos=df_codigos[df_codigos['pais']!='malo']

df_codigos

# c)

df=pd.merge(df_anio,df_codigos,on='codigo_iso')

df
```

3. Desarrolle una función `resumen_df(df)` para encontrar el total de elementos distintos y vacíos por columnas.

```python
# respuesta
def resumen_df(df):

    nombres = df.columns

    u_list=[]
    v_list=[]

    for j in nombres:

      unique=(len(df[j].unique()))
      vacio=(df[j].isnull().sum())

      if vacio!=0:
         unique=unique-1;

      u_list.append(unique)
      v_list.append(vacio)

    result = pd.DataFrame({'nombres': nombres})

    result['elementos_distintos'] = u_list
    result['elementos_vacios'] = v_list

    return result
```

```python
# retornar
resumen_df(df)
```

4. Para los paises latinoamericano, encuentre por año  el país con mayor y menor `indice`.

 * a) Mediante un ciclo _for_.
 * b) Mediante un  _groupby_.

```python
# respuesta
america = ['ARG', 'ATG', 'BLZ', 'BOL', 'BRA', 'CAN', 'CHL', 'COL', 'CRI',
       'CUB', 'DOM', 'ECU', 'GRD', 'GTM', 'GUY', 'HND', 'HTI', 'JAM',
       'MEX', 'NIC', 'PAN', 'PER', 'PRY', 'SLV', 'SUR', 'TTO', 'URY',
       'USA', 'VEN']

df_america = df[df.codigo_iso.isin(america)]

# Mediante ciclo for
# Mayor índice

anios = df['anio'].unique()

dict_mayor= dict()

for k in anios:

    df_aux = df_america.loc[lambda x: x['anio'] == k]
    max_indice = df_aux['indice'].max()
    df_aux = df_aux[df_aux['indice']==max_indice]
    paises = list(df_aux['pais'])

    new_paises=[]

    if len(paises)<1:
      paises.append('No hay datos')
      new_paises.append(paises[0])
    elif len(paises)>=1:
      new_paises.append(paises[0])
    if (np.isnan(max_indice)==True):
      max_indice=0

    dict_mayor[k]=[new_paises[0],max_indice]

sorted(dict_mayor.items())

# Menor índice
dict_menor= dict()

for k in anios:
    df_aux = df_america.loc[lambda x: x['anio'] == k]
    min_indice = df_aux['indice'].min()
    df_aux=df_aux[df_aux['indice']==min_indice]
    paises=list(df_aux['pais'])

    new_paises=[]

    if len(paises)<1:
      paises.append('No hay datos')
      new_paises.append(paises[0])
    elif len(paises)>=1:
      new_paises.append(paises[0])
    if (np.isnan(min_indice)==True):
      min_indice=0
    dict_menor[k]=[new_paises[0],min_indice]

sorted(dict_menor.items())

# Mediante un groupby
# Mayor índice

max_indice = df_america.groupby('anio').indice.max()
data_max = df_america.merge(max_indice, on='anio', suffixes=('','_max'))
data_max= data_max[data_max.indice==data_max.indice_max].drop(['indice_max','codigo_iso','ranking',], axis=1)
by_anio = data_max.sort_values('anio')

by_anio

# Menor índice

min_indice = df_america.groupby('anio').indice.min()
data_min = df_america.merge(min_indice, on='anio', suffixes=('','_min'))
data_min= data_min[data_min.indice==data_min.indice_min].drop(['indice_min','codigo_iso','ranking',], axis=1)
by_anio = data_min.sort_values('anio')

by_anio
```
